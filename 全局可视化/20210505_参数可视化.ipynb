```python
data=pd.read_excel('原始数据.xlsx')
data
```

```python
x=data['日期'].tolist()
injected=data['现存确诊'].tolist()
dead=data['死亡人数'].tolist()
recoverd=data['治愈人数'].tolist()
total=data['累计确诊'].tolist()
```

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker


c = (
    Line(init_opts=opts.InitOpts(theme=ThemeType.ROMANTIC))
    .add_xaxis(x)
    .add_yaxis("现存确诊", injected, is_smooth=True)
    .add_yaxis("死亡人数", dead, is_smooth=True)
    .add_yaxis("治愈人数", recoverd, is_smooth=True)
    .add_yaxis("累计确诊", total, is_smooth=True)
    .set_series_opts(
        areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="数据概览"),
        xaxis_opts=opts.AxisOpts(
            axistick_opts=opts.AxisTickOpts(is_align_with_label=True),
            is_scale=False,
            boundary_gap=False,
        ),
    )
    .render("line_areastyle_boundary_gap.html")
)
```
