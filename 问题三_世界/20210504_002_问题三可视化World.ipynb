```python
import pandas as pd
data=pd.read_excel("World.xlsx")
print(data[data.isnull().T.any()])
print("---------------------------------")
data
```

```python
temp0=pd.read_csv('third_predict0.csv')
temp1=pd.read_csv('third_predict1.csv')
temp2=pd.read_csv('third_predict2.csv')
temp3=pd.read_csv('third_predict3.csv')
temp4=pd.read_csv('third_predict4.csv')
temp5=pd.read_csv('third_predict5.csv')
temp6=pd.read_csv('third_predict6.csv')
temp7=pd.read_csv('third_predict7.csv')
temp8=pd.read_csv('third_predict8.csv')
temp9=pd.read_csv('third_predict9.csv')
temp10=pd.read_csv('third_predict10.csv')
temp11=pd.read_csv('third_predict11.csv')
temp12=pd.read_csv('third_predict12.csv')
temp13=pd.read_csv('third_predict13.csv')
predict=pd.concat([temp0,temp1,temp2,temp3,temp4,temp5,temp6,temp7,temp8,temp9,temp10,temp11,temp12,temp13],axis=0)
predict
```

```python
import matplotlib.pyplot as plt
x=list(range(1,data.shape[0]+1))
y1=data['累计确诊'].tolist()
y2=data['治愈人数'].tolist()
y3=data['死亡人数'].tolist()
plt.plot(x,y1,'r-',label='Injected_true',linewidth = 3)
plt.plot(x,y2,'b-',label='Recovered_true',linewidth = 3)
plt.plot(x,y3,'y-',label='Dead_true',linewidth = 3)
data=predict
x_pre=list(range(1,data.shape[0]+1))
y1_pre=data['累计确诊'].tolist()
y2_pre=data['治愈人数'].tolist()
y3_pre=data['死亡人数'].tolist()
plt.plot(x_pre,y1_pre,'c-',label='Injected_predict',linewidth = 1)
plt.plot(x_pre,y2_pre,'g-',label='Recovered_predict',linewidth = 1)
plt.plot(x_pre,y3_pre,'m-',label='Dead_predict',linewidth = 1)
plt.legend()
plt.show()
```

```python
plt.figure(1)
plt.plot(x,y1,'r-',label='Injected_true',linewidth = 3)
plt.plot(x_pre,y1_pre,'c-',label='Injected_predict',linewidth = 1)
plt.legend()
plt.figure(2)
plt.plot(x,y2,'b-',label='Recovered_true',linewidth = 3)
plt.plot(x_pre,y2_pre,'g-',label='Recovered_predict',linewidth = 1)
plt.legend()
plt.figure(3)
plt.plot(x,y3,'y-',label='Dead_true',linewidth = 3)
plt.plot(x_pre,y3_pre,'m-',label='Dead_predict',linewidth = 1)
plt.legend()
plt.show()
```
